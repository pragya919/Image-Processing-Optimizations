We read the background and the foreground images and and resize both of them to make them of equal size.

Next we create a mask by using the green pixels in the background. We basically define a range that will cover all the green pixels in the background and separate the main character from it. 

Then we add the masked image to the background image to get the required output.

The above technique is called Chroma Keying!

### Question 4 - 1 part

```python
def bitQuantizeImage(image,k):
    out = np.zeros((image.shape[0],image.shape[1]))
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            out[i][j]=math.floor(image[i][j]/pow(2, 8-k));
    return out
```

### Question 5 - 1 part

```python
def linContrastStretching(image,a,b,m1,m2):
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            for k in range(image.shape[2]):
                image[i][j][k]=a + (image[i][j][k]-m1)*((b-a)/(m2-m1))
    return image
```

#### Theoretical Explanation of Constrast Stretching

Linear Constrast Stretching is a type of normalization. If the input image has an intensity values in a particular range then by using this technique, we can widen the range to the maximum allowed intensity values. 

For this, suppose the intensity of pixel in original image is referred as i - 

The range of intensity in original image = (a,b)

The required range is = (c,d)

Then by simple maths we can convert the value of 'i' in range (a,b) to an equivalent value 'j' in range (c,d) - j = (((i-a)/(b-a)) * (d-c)) + c

To prove, put i = a, as a result j=c and similarly, put i = b, as a result j=d

```python
def piecewiseLinTransform(image,k1,k2,a,b):
    #declaring a new output array
    image1=np.copy(image)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
             for k in range(len(k1)):
                #looping over the pixels of the input image for every interval
                if(image[i][j]>=a[k] and image[i][j]<=b[k]):
                    image1[i][j] = int(k1[k]*image[i][j] + k2[k])
                    #if the pixels are in the interval then change them            
    #return the output image
    return image1

img=cv2.imread('snape1.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
t1=[5,-10,5,-10,5,-10]
t2=[0,765,-382.5,1530,-765,2295]
t3=[0,51,76.5,127.5,153,204]
t4=[51,76.5,127.5,153,204,229.5]
out = piecewiseLinTransform(img,t1 , t2, t3, t4)
plt.imshow(out,cmap='gray')
plt.show()
```

```python
def histEqualization(image):
    p = {}
    hist=np.zeros(256)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            hist[image[i][j]]+=1
    a=np.zeros(hist.shape)
    a[0]=hist[0]
    for i in range(1,hist.shape[0]):
        a[i]=a[i-1]+hist[i]
    for i in range(0,256):
        p[i] = int((a[i]-a.min())*255/(a.max()-a.min()))
    a = (a-a.min())*255/(a.max()-a.min())
    out=a[image]
    return p,out
```

```python
img=cv2.imread('fred.jpeg')
plt.subplot(2,2,1)
plt.imshow(img,cmap='gray')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.subplot(2,2,2)
plt.hist(img,bins=50)
plt.show()

_,output = histEqualization(img)
plt.subplot(2,2,3)
plt.imshow(output,cmap='gray')
plt.subplot(2,2,4)
plt.hist(output,bins=50)
plt.show()
```

```python
def find(value,dic):
    k,v= min(dic.items(), key=lambda x: abs(value - x[1]))
    return k

def histMatching(img,img1):
    p,_ = histEqualization(img)
    q,_ = histEqualization(img1)
    for i in p:
        p[i]=find(p[i],q)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            img[i][j]=p[img[i][j]]
    return img

img=cv2.imread('map.jpeg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img1=cv2.imread('map-reference.jpeg')
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img = histMatching(img,img1)
plt.subplot(1,2,1)       
plt.imshow(img,cmap='gray')
# plt.hist(img)
plt.subplot(1,2,2)       
plt.imshow(img1,cmap='gray')
# plt.hist(img1)
plt.show()
```

```python
def gammaTransform(img,gamma):
    output = np.zeros((img.shape[0],img.shape[1]))
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            output[i][j]=((img[i][j]/255)**gamma)*255
    return output

img=cv2.imread('locket.jpeg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
out = gammaTransform(img,0.3)
plt.subplot(1,2,2)
plt.imshow(out,cmap='gray')
plt.show()
```

```python
img=cv2.imread('voldemort.jpeg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
out = gammaTransform(img,2)
plt.subplot(1,2,2)
plt.imshow(out,cmap='gray')
plt.show()
```

#### Observation - 

When gamma is less than one, the image becomes brighter and when it is greater than one, the image becomes darker. This is because, when we divide the intensity values by 255 then all the values are less than or equal to one and hence by raising them to a power which is greater than one, we are basically reducing their values and then scaling it back to 255, hence the image becomes darker. Same explanation can be given for brighter images
